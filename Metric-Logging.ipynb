```python
# Setting up
from azureml.core import Experiment, Workspace, Run
import azureml.core
import numpy as np
from tqdm import tqdm

print(" The current version of SDK is: ", azureml.core.VERSION)
```

```python
# Initialize workspace
ws = Workspace.from_config()
print("Workspace name: " + ws.name, 'Azure region: ' + ws.location)
```

```python
# Set experiment
experiment = Experiment(workspace=ws, name = 'logging-api-test')
```

```python
# Start logging for the run
run = experiment.start_logging()

# access the run id
run_id = run.id

# logging different values for the variables
scale_factor = 3
category = 'yellow'
```

```python
# logging string metrics
run.log(name ='category', value = category)
```

```python
#log numerical values
run.log(name='scale factor', value = scale_factor)
run.log(name='magic Number', value = 42 * scale_factor)
```

```python
run.complete()
```

```python
# Analysing results
fetched_run = Run(experiment, run_id)
fetched_run
```

```python
# Getting all metrics for a run
run.get_metrics()
```

```python
# Getting a particular metric
fetched_run.get_metrics(name = 'scale factor')
```

```python
fetched_run.get_file_names()
```

```python
# tagging a run
fetched_run.tag('My favorite run')
fetched_run.tag('Competition rank', 1)

fetched_run.get_tags()
```
